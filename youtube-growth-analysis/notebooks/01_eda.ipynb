# Exploratory Data Analysis (EDA) on YouTube Video Metadata

In this notebook, we will perform exploratory data analysis on the YouTube video metadata to uncover trends and patterns related to video engagement.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualisation style
sns.set(style='darkgrid')
```

```python
# Load the dataset
data_path ='../data/sample_videos.csv'
df = pd.read_csv(data_path )

# Display the first few rows of the dataframe
df.head()
```

```python
# Summary statistics
df.describe(include='all')
```

```python
# Check for missing values
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
```

```python
# Visualize the relationship between title length and views
df['title_length'] = df['video_title'].apply(len)

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='title_length', y='views')
plt.title('Title Length vs Views')
plt.xlabel('Title Length')
plt.ylabel('Number of Views')
plt.savefig('../output/plots/title_length_vs_views.png')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df,x="views",y="likes")
plt.xlabel('Views')
plt.ylabel('Likes')
plt.title('Views vs Likes')
plt.savefig('../output/plots/views_vs_likes.png')
plt.show()
```
